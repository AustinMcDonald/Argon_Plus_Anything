```python
import numpy as np 
from tables import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
APA = open_file("xenon.h5")
```

```python
Percent = ["15.0", "10.0", "0.0001"]
```

```python
APA.root
```

```python
APA.root.HELIUM3
```

```python
plt.figure(figsize=(7,7))
plt.title("drift")
for x in range(0,len(Percent)):
    Per = Percent[x]
    redu = APA.root.HELIUM3[Per].Reduced_Field.read()
    dz = APA.root.HELIUM3[Per].Drift_Velocity.read()
    plt.plot(redu,dz,linewidth=2,alpha=0.7, label=Per)
plt.legend(loc="lower right",fontsize=12)
plt.show()

plt.figure(figsize=(7,7))
plt.title("Dl")
for x in range(0,len(Percent)):
    Per = Percent[x]
    redu = APA.root.HELIUM3[Per].Reduced_Field.read()
    dz = APA.root.HELIUM3[Per].Long.read()
    plt.plot(redu,dz,linewidth=2,alpha=0.7, label=Per)
plt.legend(loc="lower right",fontsize=12)
plt.show()

plt.figure(figsize=(7,7))
plt.title("Dt")
for x in range(0,len(Percent)):
    Per = Percent[x]
    redu = APA.root.HELIUM3[Per].Reduced_Field.read()
    dz = APA.root.HELIUM3[Per].Trans.read()
    plt.plot(redu,dz,linewidth=2,alpha=0.7, label=Per)
plt.legend(loc="lower right",fontsize=12)
plt.show()
```
